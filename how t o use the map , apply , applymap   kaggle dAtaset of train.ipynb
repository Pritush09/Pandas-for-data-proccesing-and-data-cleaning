```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("http://bit.ly/kaggletrain")
```

```python
train.head()
```

```python
# map is a series method 
train['Sex_num'] = train.Sex.map({'female':0, "male":1})
```

```python
train.head()
```

```python
# apply method
# it apply a function in each element of the series
train['Name_lenght'] = train.Name.apply(len)
```

```python
train.head()
```

```python
train.drop("train_lenght",axis=1)
```

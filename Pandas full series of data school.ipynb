# how do filter rows of a pandas dataframe by column values 

```python
imdb2 = pd.read_csv("http://bit.ly/imdbratings")
```

```python
imdb2.head()
```

```python
imdb2.shape
```

```python
# filtering the data on thee basis of the movies that are above 200 min 
```

```python
# fisrtly to understand the direct method present in pandas we will break it into parts
```

```python
boolean = []
for lenght in imdb2.duration:
    if lenght >= 200:
        boolean.append(True)
    else:
        boolean.append(False)
```

```python
boolean[0:5]
```

```python
len(boolean)
```

```python
# now we convert the boolean list into the pandas series 
is_long = pd.Series(boolean)
```

```python
# we can use this to remove our foor loop 
is_long= imdb2.duration >=200
is_long.head()
```

```python
is_long.head()
```

```python
# now we will pass the boolean series in the imdb2 so that it will return the part which we want
imdb2[is_long]
```

```python
# it also work exactly the same
imdb2[imdb2.duration >=200]
```

# Bonus Tip

```python
# for getting the column out of our created dataframe
imdb2[imdb2.duration >=200].genre
```

```python
# final point usage of loc it usually is the best method now days 2021 and powerfull
imdb2.loc[imdb2.duration >=200,'genre']
```

# how to filter the dataframe by multiple filter criteria

```python
imdb3 = pd.read_csv("http://bit.ly/imdbratings")
```

```python
# u have to keep the order in parenthesis it the filteered data which we wanted to get 
# we have to use the & == and operator
imdb3[(imdb3.duration >= 200) & (imdb3.genre == 'Drama')]
```

```python
# for the the or conditon e use | 
# it will simply return the data which satisfy the one of the given condition 
imdb3[(imdb3.duration >= 200) | (imdb3.genre == 'Drama')]
```

# Bonnu Tip

```python
imdb3[(imdb3.duration == "Action") | (imdb3.genre == 'Drama') | (imdb3.genre == 'Crime')]
```

```python
# to simply shorten the above statment 
imdb3[imdb3.genre.isin(['Crime','Action',"Drama"])] # it is a function 
```

# how to handel missing values in panda dataset

```python
ufos = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufos.tail()
```

```python
ufos.isnull().tail() # it detect the null values and return true for a null values
```

```python
ufos.notnull().tail()
```

```python
ufos.isnull().sum() # here pandas is converting true to 1 and false to 0
# and the mathamatical operation the axis 0 defines the column remember
```

```python
# we canpass the same thing like this to understand the null values of a particular given series 
ufos[ufos.City.isnull()]
```

# what do with this missing values 

```python
ufos.shape
```

```python
# Firstly we can simply drop the row in a givem dataset
# droping if any of its values are missing 
ufos.dropna(how='any').shape #it has inplace parameter as false as it does not directly changes the dataset
```

```python
ufos.shape
```

```python
# Secondly drop if only all of its valuers are missing 
ufos.dropna(how='all').shape
```

```python
# we can use a list of column whcich we wanna consider 
ufos.dropna(subset= ['City','Shape Reported'],how= 'any').shape
# we use above to drop the row if any of the given parameter is missing 
```

```python
ufos.dropna(subset= ['City','Shape Reported'],how= 'all').shape
# only drop if all of the given parameter is missing 
```

# bonus tip 

```python
ufos["Shape Reported"].fillna(value='VARIOUS',inplace=True)
```

```python
# filling the missing values
ufos['Shape Reported'].value_counts(dropna=False) # we use dropna in this so too see how many missing 
#values are there in datset
```
